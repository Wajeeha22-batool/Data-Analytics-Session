### OOP (Object Oriented Programming)

OOP is a programming concept of objects, which can contain data and code:

- Data in the form of fields (often called attributes or properties)

- Code in the form of procedures (often called methods)

```python
class Person: #Person is class name
    def say_hello(): #method of class Person
        print("Hello!")

```

```python
#object
Person()
```

```python
p = Person()
p.say_hello() #object as argument
```

```python
class Person: #Person is class name
    def say_hello(self): #method of class Person
        print("Hello!")

p = Person()
p.say_hello() #method calling of class Person
```

```python
class Person: #Person is class name
    def say_hello(self): #method of class Person
        print(f"Hello! {self}")

p = Person()
p.say_hello()
```

```python
class Person: #Person is class name
    def say_hello(self): #1st method of class Person
        print("Hello!")

    #2nd method
    def say_goodbye(x,y):
        print("Goodbye!")
```

```python
p = Person()
p
```

```python
p.say_goodbye() #calling method of class
```

```python
class Car:
    def __init__(self, color, brand):
        self.color = color    # attribute
        self.brand = brand    # attribute

    def start_engine(self):   # 1st method
        print(f"The {self.brand} car's engine has started.")

    def describe(self,y):       #2nd method
        print(f"This car is a {self.color} {self.brand}.")

    #model
    
    #cartype
```

```python
my_car = Car('black','Honda')
```
