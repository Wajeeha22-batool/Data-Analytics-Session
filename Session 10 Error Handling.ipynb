```python
try:
    x = int(input("enter num1:"))
    y = int(input("enter num2:"))

    result = x/y
    print(result)

except ZeroDivisionError:
    print('cannot divided by zero')
except ValueError:
    print('invalid integer')
```

```python
#try
#except
#else
try:
    x = int(input("enter num1:"))
    y = int(input("enter num2:"))

    result = x/y
    
except ZeroDivisionError:
    print('cannot divided by zero')
except ValueError:
    print('invalid integer')
else: #optional
    print(result)
```

```python
try:
    x = int(input("enter num1:"))
    y = int(input("enter num2:"))

    result = x/y
    print(result)

except ZeroDivisionError:
    print('cannot divided by zero')
except TypeError:
    print('invalid integer')
except:
    print('unknown error')
```

```python
#try
#except
#else
#finally
try:
    x = int(input("enter num1:"))
    y = int(input("enter num2:"))

    result = x/y
    
except ZeroDivisionError:
    print('cannot divided by zero')
except ValueError:
    print('invalid integer')
else: #optional
    print(result)
finally:
    print('program executed')
```
