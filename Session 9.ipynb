```python
# Arbitary number positional arguments
def grocery(*args):
    print(args)

grocery('salt','ghee','sugar','rice') #positional arguments
```

```python
# Arbitary number positional arguments
def grocery1(*args):
    print(xyz)

grocery1('salt','sugar')#positional arguments
```

```python
# Arbitary number positional arguments
def grocery1(*xyz):
    print(xyz)

grocery1('salt','sugar')#positional arguments
```

```python
# Arbitary number of keyword arguments

def bio(**kwargs):
    print(kwargs)

bio(name='ali',age=22,gender='male')
```

```python
def bio(**abc):
    print(abc)

bio(name='ali',age=22,gender='male')
```

### Nested function

```python
def add(a,b):
    print(a+b)
```

```python
def subtract(a,b):
    print(a-b)
```

```python
def calculate(a, b):
    def add(x, y): 
        return x + y
    def subtract(x, y):
        return x - y
    
    print("Addition:", add(a, b))
    print("Subtraction:", subtract(a, b))

calculate(1, 4)
```
